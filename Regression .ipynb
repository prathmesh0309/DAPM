```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
import numpy as np
import operator

file_path = '/Users/prathameshmandhare/Downloads/wildfire.csv'  # Update with your file path
wildfire_data = pd.read_csv(file_path)

# Assuming 'brightness' as independent variable and 'frp' as dependent variable
X = wildfire_data[['brightness']].values
y = wildfire_data['frp'].values

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_lin = lin_reg.predict(X)

# Polynomial Regression (Degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
y_pred_poly = poly_reg.predict(X_poly)

# Plotting both Linear and Polynomial Regression
plt.figure(figsize=(14, 6))

# Linear Regression Plot
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred_lin, color='red')
plt.title('Linear Regression')
plt.xlabel('Brightness')
plt.ylabel('FRP')

# Polynomial Regression Plot
plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue')
# Sorting X_poly and y_pred_poly for a smooth curve
sorted_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(X,y_pred_poly), key=sorted_axis)
X_sorted, y_pred_poly_sorted = zip(*sorted_zip)
plt.plot(X_sorted, y_pred_poly_sorted, color='red')
plt.title('Polynomial Regression')
plt.xlabel('Brightness')
plt.ylabel('FRP')

plt.tight_layout()
plt.show()
```
